```python
#Process the final term frequency data, 

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
#Read in
colnames = ['Character', 'Word', 'TF-IDF']
tfidf_data = pd.read_csv(r"data\tf_idf_full.csv", names=colnames, encoding='latin1')

colnames = ['Character', 'Word', 'Frequency']
freq_data = pd.read_csv(r"data\term_frequencies.csv", names=colnames, encoding='latin1')
```

```python
#The list of characters to create plots for
characters = ["Leslie Knope", "Ben Wyatt", "Tom Haverford", "Ron Swanson", "April Ludgate", "Andy Dwyer", "Chris Traeger", "Ann Perkins", "Jerry Gergich", "Donna Meagle"]
```

```python
def rand_colors(n):
    output = []
    for i in range(n):
        color = [random.randint(0, 255)/255, random.randint(0, 255)/255, random.randint(0, 255)/255]
        output.append(color)
    return output
```

```python
#Plots for TF-IDF values
c = 0
fig, axes = plt.subplots(nrows=5, ncols=2)
for row in axes:
    for col in row:
        print(characters[c])
        d = tfidf_data[tfidf_data.Character == characters[c]]
        top = d.nlargest(15, 'TF-IDF')
        p = top.plot(x='Word', y='TF-IDF', kind='barh', ax = col)
        p.invert_yaxis()
        p.set_title(characters[c])
        p.get_legend().remove()
        c+=1
```

```python
#Plot for word count by episode
avg_words_per_season = pd.read_csv(r'data/term_frequencies_by_season.csv', encoding='latin1')
print(avg_words_per_season)
```

```python
bottom_margin = [0,0,0,0,0,0,0]

fig, axes = plt.subplots()
plt.sca(axes)

colors = rand_colors(10)
for n, c in enumerate(characters):
    print(c)
    val = (avg_words_per_season.loc[c])
    ind = np.arange(1,8)
    plt.bar(ind, val, bottom=bottom_margin, color = colors[n])
    bottom_margin = [sum(i) for i in zip(bottom_margin, val)] #Add values to bottom margin for subsequent char
    print(bottom_margin)
```

```python
def bar_plot_avg_words():
    data = pd.read_csv(r'data/term_frequencies_by_season.csv', encoding='latin1')
    width = .1
    curr_width = 0
    colors = rand_colors(10)
    for c in characters:
        char_row = data.loc[data.Character == c]
        char_row = char_row.astype(str)
        char_row = char_row.values.tolist()[0]
        char_row.pop()
        season = np.arange(1,8)
        plt.bar(x=season+curr_width, height=char_row, width=width)
        curr_width += width
bar_plot_avg_words()
```
